```python
import pandas as pd
import os, time
from time import sleep
```

### Settings

```python
#Python version
python=3.7

# ForecastBuilder path
dataDirectory='/home/CarterHumphreys/ForecastBuilder'

# Forecast Setup
startTime='2020-01-29 12:00'  # Forecast start time in local time (YYYY-MM-dd HH:mm (24 hour format ONLY))
duration=15  # Forecast length in hours
update=12  # Hours until next update from start time

# Models to use
useHRRR=True
useNAM=True
useGFS=False
```

### Initialization

```python
# Read in forecast points
forecastPoints=pd.read_csv(f'{dataDirectory}/bin/forecast_points.csv')
```

### Script Builder

```python
# Move to module directory
print(f'cd {dataDirectory}/modules')

# Generate latest run times
if useHRRR:
    print(f"""python{python} -c 'import GetLatest; GetLatest.getLatestRun("hrrr")'""")

if useNAM:
    print(f"""python{python} -c 'import GetLatest; GetLatest.getLatestRun("nam1hr")'""")
    
if useGFS:
    print(f"""python{python} -c 'import GetLatest; GetLatest.getLatestRun("gfs_0p25_1hr")'""")
    
# Generate model data
for index, row in forecastPoints.iterrows():
    if useHRRR:
        print(f"""python3.7 -c 'import GetModelData; GetModelData.getData("{row['PointID']}", "hrrr", "{startTime}", {duration})' &""")
    if useNAM:
        print(f"""python3.7 -c 'import GetModelData; GetModelData.getData("{row['PointID']}", "nam1hr", "{startTime}", {duration})' &""")
    if useGFS:
        print(f"""python3.7 -c 'import GetModelData; GetModelData.getData("{row['PointID']}", "gfs_0p25_1hr", "{startTime}", {duration})' &""")
```
