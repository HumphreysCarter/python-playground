```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
df=pd.read_csv('http://carterhumphreys.com/raspberry-pi/pi5/RasPi5.dat', parse_dates=['DateTime'])
df.dtypes
```

```python
fig, axs = plt.subplots(2, 2, figsize=(18, 10))
axs[0, 0].plot(df['DateTime'], df['CPU Temp'], 'tab:blue')
axs[0, 0].set_title('CPU Temperature')
axs[0, 0].set_ylabel('degC')

axs[0, 1].plot(df['DateTime'], df['CPU Usage'], 'tab:orange')
axs[0, 1].set_title('CPU Usage')
axs[0, 1].set_ylabel('%')

axs[1, 0].plot(df['DateTime'], df['RAM Usage'], 'tab:green')
axs[1, 0].set_title('RAM Usage')
axs[1, 0].set_ylabel('Mb')

axs[1, 1].plot(df['DateTime'], df['Disk Usage'], 'tab:red')
axs[1, 1].set_title('Disk Usage')
axs[1, 1].set_ylabel('%')

for ax in fig.get_axes():
    ax.set_xlim(datetime.utcnow()-timedelta(hours=2), datetime.utcnow())

plt.suptitle(f'Pi 5 - Raspberry Pi 4B', fontsize=16, y=0.95)

plt.show()
```
