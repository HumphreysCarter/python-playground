```python
import os
from datetime import datetime
```

```python
def find_nthIndex(fullString, find, n):
    start = fullString.find(find)
    while start >= 0 and n > 1:
        start = fullString.find(find, start+len(find))
        n -= 1
    return start
```

```python
dir='/home/CarterHumphreys/research/les_neural_network/data/BandPosition'

for i in range(0, 36):
    subDir=f'{dir}/Ontario_LES_Event{str(i+1).zfill(2)}'
    
    for file in os.listdir(subDir):
        try:
            timeStr=file[(find_nthIndex(file, '_', 2)+1):find_nthIndex(file, '.csv', 1)]
            dateTime=datetime.strptime(timeStr, '%m%d%Y_%H%M%S')
            newFile = file.replace(timeStr, dateTime.strftime('%Y%m%d_%H%M%S'))
        
            os.system(f'mv {subDir}/{file} {subDir}/{newFile}')
        except:
            print(f'Skipping Ontario_LES_Event{str(i+1).zfill(2)}/{file}')
```
